```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(color_codes = True)
%matplotlib inline
```

```python
auto = pd.read_csv('data/Automobile.csv')
auto.head()
```

```python
sns.distplot(auto.highway_mpg);
```

```python
sns.histplot(auto.highway_mpg, kde=True);
```

```python
sns.jointplot(data = auto,x = 'engine_size', y= 'horsepower');
```

```python
sns.jointplot(data = auto,x = 'engine_size', y= 'horsepower', kind='hex');
```

```python
sns.jointplot(data = auto,x = 'engine_size', y= 'horsepower', kind = "kde");
```

```python
sns.jointplot(data = auto,x = 'engine_size', y= 'horsepower', kind = "kde", fill=True);
```

```python
sns.pairplot(data=auto[['normalized_losses', 'engine_size', 'horsepower']]);
```
