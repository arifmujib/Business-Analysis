```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = 0
```

```python
df +=2
a = stats.t.rvs(df, size=10000)
b = stats.norm.rvs(size=10000)
ax = plt.figure(figsize=(10,6))
plt.title('For degree of freedom %s'%df, size=24)
sns.kdeplot(a, color='b')
sns.kdeplot(b, color='r')
plt.legend(['t-dist', 'Normal distribution'], fontsize=14)
plt.xlim(-10,10)
```
