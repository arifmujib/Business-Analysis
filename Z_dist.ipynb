# New random variable is the average length of the bundle.
# $Y = \frac{X_1 + X_2 + ....X_6}{6}$

# Mean: $E[Y] = E[\frac{X_1 + X_2 + ....X_6}{6}] = 125.7$

# Variance: $V[Y] = V[\frac{X_1 + X_2 + ....X_6}{6}] = \frac{1}{6^2}*(V[X_1 + X_2 + ....X_6]) = \frac{6*1.5}{6^2}=\frac{1.5}{6} = 0.25$
# SD = $\sqrt(Variance) = 0.5$

```python
a = np.random.normal(size=100000, loc= 125.7, scale=.5)
sns.histplot(a, kde=True)
```

# $P(124.1 < Y < 124.1)$
# = $P(\frac{124.1 - 125.7}{0.5} < \frac{Y - 125.7}{0.5} < \frac{124.7 - 125.7}{0.5})$
# = $P( -3.2 < Z < -2)$

```python
from scipy import stats
```

```python
ans = stats.norm.cdf(124.7, loc= 125.7, scale=.5) -  stats.norm.cdf(124.1, loc= 125.7, scale=.5)
print(ans)
```

```python
ans = stats.norm.cdf(-2) -  stats.norm.cdf(-3.2)
print(ans)
```
