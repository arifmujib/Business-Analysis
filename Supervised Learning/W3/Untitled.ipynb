```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

```python
data = pd.read_csv('anime_data.csv')
data.head()
```

```python
data.sample(5)
```

```python
data.shape
```

```python
df = data.copy()
```

```python
df.info()
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
df.describe(include='all').T
```

```python
df.dropna(subset=['rating'], inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)[:10]
```

```python
df[df.startYr.isnull()]
```

```python
df.dropna(subset=['startYr'], inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)[:10]
```

```python
df[df.finishYr.isnull()]
```

```python
df[df.finishYr.isnull()][['startYr', 'ongoing']].describe(include='all')
```

```python
plt.hist(df[df.finishYr.isnull()].startYr)
```

```python
df.finishYr.fillna(2020, inplace=True)
df.isnull().sum().sort_values(ascending=False)[:10]
```

```python
df['years_running'] = df.finishYr - df.startYr
df.drop(columns=['finishYr','startYr'], inplace=True)
df.head()
```

```python
df.sznOfRelease.fillna('is_missing', inplace=True)
df.isnull().sum().sort_values(ascending=False)[:10]
```

```python
df.mediaType.value_counts()
```

```python
df.mediaType.fillna('Other', inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)[:10]
```

```python
df.mediaType.value_counts()
```

```python
df1 = df.copy()
 df1.groupby(['studio_primary', 'mediaType'])[['duration', 'watched']].transform(lambda x: x.fillna(x.median()))
```

```python
df1[['duration', 'watched']] = df1.groupby(['studio_primary', 'mediaType'])[['duration', 'watched']].transform(lambda x: x.fillna(x.median()))
```

```python
df1.isnull().sum().sort_values(ascending=False)[:8]
```

```python
df1.drop(columns=['description', 'title'], inplace=True)
df1.describe(include="all").T
```

```python
df1["duration"].fillna(df1.duration.median(), inplace=True)
df1.isnull().sum().sort_values(ascending=False)[:8]
```
